```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the CSV file
file_path = '/content/drive/MyDrive/Betathalassemia_main.csv'  # Update this with your file path
data = pd.read_csv(file_path)
```

```python
data.dropna(subset=['Age', 'Sex', 'Family History', 'Phenotype'], inplace=True)
```

```python
bins = [4, 11, 19, 27, 39, 46]  # Define your age bins
labels = ['4-10', '11-18', '19-26', '27-38', '39-45']  # Labels for the bins
data['Age_Cluster'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)
```

```python
summary = data.groupby(['Sex', 'Age_Cluster', 'Family History', 'Phenotype']).size().unstack(fill_value=0)
```

```python
# Plotting Family History by Sex and Age Cluster
plt.figure(figsize=(14, 7))
sns.countplot(x='Age_Cluster', hue='Sex', data=data[data['Family History'] == 'Yes'])
plt.title('Family History by Sex and Age Cluster')
plt.xlabel('Age Cluster')
plt.ylabel('Count')
plt.legend(title='Sex')
plt.show()
```

```python
# Plotting Phenotype by Sex and Age Cluster
plt.figure(figsize=(14, 7))
sns.countplot(x='Phenotype', hue='Sex', data=data, palette="muted")
plt.title('Phenotype by Sex')
plt.xlabel('Phenotype')
plt.ylabel('Count')
plt.legend(title='Sex')
plt.show()
```

```python
# Plotting Family History and Phenotype
plt.figure(figsize=(14, 7))
sns.countplot(x='Phenotype', hue='Family History', data=data, palette="coolwarm")
plt.title('Phenotype by Family History')
plt.xlabel('Phenotype')
plt.ylabel('Count')
plt.legend(title='Family History')
plt.show()
```

```python
# Plotting Family History and Phenotype
plt.figure(figsize=(14, 7))
sns.countplot(x='Phenotype', hue='Sex', data=data, palette="coolwarm")
plt.title('Phenotype by Sex')
plt.xlabel('Phenotype')
plt.ylabel('Count')
plt.legend(title='Sex')
plt.show()
```

```python
# Plotting Family History and Phenotype
plt.figure(figsize=(14, 7))
sns.countplot(x='Phenotype', hue='Age_Cluster', data=data, palette="coolwarm")
plt.title('Phenotype by Age')
plt.xlabel('Phenotype')
plt.ylabel('Count')
plt.legend(title='Age_Cluster')
plt.show()
```
